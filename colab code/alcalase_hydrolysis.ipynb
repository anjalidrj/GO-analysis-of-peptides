```python
#converts the genbank file to a fasta with only the protein sequences
from Bio import SeqIO

gbk_filename = list(uploaded.keys())[0]

with open(gbk_filename) as f, open("proteome.fasta", "w") as out:
    for record in SeqIO.parse(f, "genbank"):
        for feature in record.features:
            if feature.type == "CDS":
                protein_id = feature.qualifiers.get("protein_id", ["unknown"])[0]
                product = feature.qualifiers.get("product", ["unknown product"])[0]
                translation = feature.qualifiers.get("translation", [None])[0]
                if translation:
                    out.write(f">{protein_id} {product}\n{translation}\n")
```

```python
#cleaves the proteome using alcalase, allowing 0, 1 and 2 missed cleaves.
#for this to work the proteome FASTA file MUST be named "proteome.fasta"
#don't run the next cell till this shows "done"
import re
from Bio import SeqIO

def alcalase_digest(sequence, miss_cleavages=0, min_len=6):
    pattern = r'(?<=[EMLYKKQ])'
    sites = [m.end() for m in re.finditer(pattern, sequence)]
    sites = [0] + sites + [len(sequence)]
    base_peptides = [sequence[sites[i]:sites[i+1]] for i in range(len(sites)-1)]
    peptides = []
    for i in range(len(base_peptides)):
        for mc in range(miss_cleavages + 1):
            if i + mc < len(base_peptides):
                pep = "".join(base_peptides[i:i+mc+1])
                if len(pep) >= min_len:
                    peptides.append((pep, mc))
    return peptides

with open("proteome_B.fasta") as f, \
     open("peptides_mc0.fasta", "w") as out0, \
     open("peptides_mc1.fasta", "w") as out1, \
     open("peptides_mc2.fasta", "w") as out2:

    for record in SeqIO.parse(f, "fasta-pearson"):
        peptides = alcalase_digest(str(record.seq), miss_cleavages=2)
        for i, (pep, mc) in enumerate(peptides):
            if mc == 0:
                out0.write(f">{record.id}_pep{i+1}_mc0\n{pep}\n")
            elif mc == 1:
                out1.write(f">{record.id}_pep{i+1}_mc1\n{pep}\n")
            elif mc == 2:
                out2.write(f">{record.id}_pep{i+1}_mc2\n{pep}\n")

print("done")
```
